```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
len(tokenizer.word_counts)
```

```python
tokenizer.fit_on_texts([jokes])
```

```python
tokenizer.word_index
```

```python
input_sequences  = []

for sent in jokes.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sent])[0]
    
    for i in range(1, len(tokenized_sentence)):
        n_gram = tokenized_sentence[:i+1]
        input_sequences.append(n_gram)
```

```python
input_sequences
```

```python
# apply 0 padding
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
```

```python
padded_sequences
```

```python
X = padded_sequences[:, :-1]
y = padded_sequences[:, -1]
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=1893)
```

## Modelling LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()

model.add(Embedding(input_dim=1893, output_dim=128, input_length=13))
model.add(LSTM(300))
model.add(Dense(1893, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=100)
```

```python
model.weights
```

## User Testing

```python
import time
import numpy as np

txt = 'i thought that '


for i in range(10):
    token_text = tokenizer.texts_to_sequences([txt])[0]
    padded_text = pad_sequences([token_text], maxlen=13, padding='pre')
    pos = np.argmax(model.predict(padded_text))
    
    for word, index in tokenizer.word_index.items():
        if index == pos:
            txt += " "+word
            print(txt)
            time.sleep(2)
```
